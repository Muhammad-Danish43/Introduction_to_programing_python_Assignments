**In this Assignment i wrote a simple python code according to assignment 3 instractions**

```python
import csv

#Read Load Data
loads = []
with open('Assignment 3 load.csv', newline='') as f:
    reader = csv.reader(f)
    next(reader)  # skip header
    for row in reader:
        if row and row[0].strip() != "":
            loads.append(float(row[0].strip()))

#Read Price Data
prices = []
with open('Assignment 3 price.csv', newline='') as f:
    reader = csv.reader(f)
    next(reader)  # skip header
    for row in reader:
        if row and row[0].strip() != "":
            prices.append(float(row[0].strip()))

# Hours
hours = list(range(1, len(loads) + 1))

# Bill Calculation
def calc_bills(loads, prices):
    hourly = [(l / 1000) * p for l, p in zip(loads, prices)]
    total = sum(hourly)
    return hourly, total

hourly_bills, total_bill = calc_bills(loads, prices)

# Output
print("{:<4} {:<8} {:<6} {:<8}".format("Hr", "Load", "KWh", "Rs"))
for h, l, c in zip(hours, loads, hourly_bills):
    print("{:<4} {:<8.2f} {:<6.3f} {:<8.2f}".format(h, l, l/1000, c))

print(f"\nTotal Bill: {total_bill:.2f} Rs")

```
